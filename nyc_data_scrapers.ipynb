```python
import pandas as pd
import json
import requests
import urllib
from urllib.request import urlopen
import re
```

# Asylum Seekers in City Care by Managing Agency (Individuals)

```python
asylum_seekers_by_agency_file = str(urlopen("https://docs.google.com/spreadsheets/u/1/d/1z2OsN8_ucCY5zhuuwJQKf0iZLDSrsPw2Wf4-KtlykTA/gviz/tq?sheet=InCityCare&headers=1&tq=select+*+&tqx=reqId:2").read())
file_contents = re.search(r'\{.*\}', asylum_seekers_by_agency_file).group()
file_contents
```

```python
asylum_seekers_by_agency_data = json.loads(file_contents)
```

```python
columns = [col["label"] for col in asylum_seekers_by_agency_data["table"]["cols"]]

# Extract rows
rows = []
for row in asylum_seekers_by_agency_data["table"]["rows"]:
    values = [cell["v"] if cell else None for cell in row["c"]]
    rows.append(values)

# Create pandas DataFrame
asylum_seekers_by_agency_df = pd.DataFrame(rows, columns=columns)
```

```python
asylum_seekers_by_agency_df
```

```python
asylum_seekers_by_agency_df.to_csv("/data/NYC_asylum_seekers_by_agency.csv")
```
