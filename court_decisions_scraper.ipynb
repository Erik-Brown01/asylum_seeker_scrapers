```python
import json
from urllib.request import urlopen
import pandas as pd
```

```python
court_decisions = pd.DataFrame()

for i in range(1,73):
    urls = [f"https://trac.syr.edu/phptools/immigration/asylum/graph.php?stat=count&timescale=fymon&base_city={i}&represented=All&denial2=1&timeunit=number",
        f"https://trac.syr.edu/phptools/immigration/asylum/graph.php?stat=count&timescale=fymon&base_city={i}&represented=All&denial2=2&timeunit=number",
        f"https://trac.syr.edu/phptools/immigration/asylum/graph.php?stat=count&timescale=fymon&base_city={i}&represented=All&denial2=3&timeunit=number"]

    for url in urls:
        response = urlopen(url)
        
        data_json = json.loads(response.read())
        # print(data_json['title'])

        df = pd.DataFrame()

        for date in data_json['timeline']:
            df = pd.concat([pd.DataFrame(date, index=[0]),df], ignore_index=True)

        city, decision = data_json['title'].rsplit(' - ', 1)

        df['city'] = city
        df['decision'] = decision

        court_decisions = pd.concat([court_decisions,df])
```
